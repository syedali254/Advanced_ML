```python
import pandas as pd
import numpy as np
import xgboost as xgb
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OrdinalEncoder
from sklearn.impute import SimpleImputer
from sklearn.model_selection import TimeSeriesSplit, cross_val_score
from sklearn.metrics import average_precision_score, classification_report
from sklearn.feature_selection import SelectKBest, f_classif

# --- Step 1: Sort by time (mandatory for temporal data) ---
df = df.sort_values('txn_timestamp').reset_index(drop=True)

# --- Step 2: Remove leaky features ---
leaky_cols = ['account_frozen', 'reversal_issued', 'dispute_filed', 'chargeback_amount']
df = df.drop(columns=[c for c in leaky_cols if c in df.columns])

# --- Step 3: Feature engineering (time-aware) ---
df['hour'] = pd.to_datetime(df['txn_timestamp']).dt.hour
df['hour_sin'] = np.sin(2 * np.pi * df['hour'] / 24)
df['hour_cos'] = np.cos(2 * np.pi * df['hour'] / 24)

# Time-aware merchant fraud rate
df = df.sort_values('txn_timestamp')
df['merchant_fraud_rate'] = (
    df.groupby('merchant_id')['is_fraud']
      .transform(lambda x: x.shift(1).expanding().mean())
      .fillna(df['is_fraud'].mean())
)

# --- Step 4: Define feature groups ---
numeric_feats = ['amount', 'hour_sin', 'hour_cos', 'merchant_fraud_rate', 
                  'user_account_age_days', 'txn_count_24h']
categorical_feats = ['merchant_category', 'card_type', 'device_type']

# Drop raw IDs
X = df[numeric_feats + categorical_feats].copy()
y = df['is_fraud'].copy()

# --- Step 5: Build leak-free preprocessing pipeline ---
numeric_transformer = Pipeline([
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler',  StandardScaler()),
])

categorical_transformer = Pipeline([
    ('imputer', SimpleImputer(strategy='constant', fill_value='unknown')),
    ('encoder', OrdinalEncoder(
        handle_unknown='use_encoded_value', unknown_value=-1
    )),
])

preprocessor = ColumnTransformer(transformers=[
    ('num', numeric_transformer, numeric_feats),
    ('cat', categorical_transformer, categorical_feats),
])

# --- Step 6: Full pipeline ---
full_pipe = Pipeline([
    ('preprocessor', preprocessor),
    ('selector', SelectKBest(f_classif, k=15)),
    ('clf', xgb.XGBClassifier(
        n_estimators=500,
        learning_rate=0.05,
        max_depth=4,
        subsample=0.8,
        colsample_bytree=0.8,
        scale_pos_weight=99,
        eval_metric='aucpr',
        random_state=42
    ))
])

# --- Step 7: Time-aware cross-validation ---
tscv = TimeSeriesSplit(n_splits=5, gap=500)
scores = cross_val_score(full_pipe, X, y, cv=tscv, scoring='average_precision')

print(f"PR-AUC per fold: {scores}")
print(f"Mean PR-AUC: {scores.mean():.4f} ± {scores.std():.4f}")

# --- Step 8: Final train/test split (time-based, NOT random) ---
split_point = int(len(X) * 0.8)
X_train, X_test = X.iloc[:split_point], X.iloc[split_point:]
y_train, y_test = y.iloc[:split_point], y.iloc[split_point:]

full_pipe.fit(X_train, y_train)
y_proba = full_pipe.predict_proba(X_test)[:, 1]
y_pred  = full_pipe.predict(X_test)

print(f"\nFinal Test PR-AUC: {average_precision_score(y_test, y_proba):.4f}")
print(classification_report(y_test, y_pred, digits=4))
```
