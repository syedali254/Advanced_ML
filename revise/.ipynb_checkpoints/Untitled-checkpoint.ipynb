```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split, StratifiedKFold, cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import (
    classification_report, confusion_matrix,
    roc_auc_score, average_precision_score,
    precision_recall_curve, f1_score
)
from sklearn.utils.class_weight import compute_class_weight
from imblearn.pipeline import Pipeline as ImbPipeline
from imblearn.over_sampling import SMOTE, ADASYN, BorderlineSMOTE
from imblearn.under_sampling import RandomUnderSampler
from imblearn.combine import SMOTETomek

# ================================================
# STEP 1 — Generate severely imbalanced fraud data
# ================================================
X, y = make_classification(
    n_samples=100_000,
    n_features=20,
    n_informative=10,
    weights=[0.999, 0.001],   # 0.1% fraud
    random_state=42
)
print(f"Total samples : {len(y):,}")
print(f"Fraud cases   : {y.sum():,}")
print(f"Fraud rate    : {y.mean()*100:.3f}%")

X_train, X_val, y_train, y_val = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)

scaler = StandardScaler()
X_tr_s = scaler.fit_transform(X_train)
X_v_s  = scaler.transform(X_val)

# ================================================
# STEP 2 — The naive model — witness the lie
# ================================================
naive = LogisticRegression(max_iter=1000)
naive.fit(X_tr_s, y_train)
y_pred_naive  = naive.predict(X_v_s)
y_proba_naive = naive.predict_proba(X_v_s)[:, 1]

print("\n=== NAIVE MODEL (no imbalance handling) ===")
print(f"Accuracy  : {(y_pred_naive == y_val).mean():.4f}")
print(classification_report(y_val, y_pred_naive, digits=4))
tn, fp, fn, tp = confusion_matrix(y_val, y_pred_naive).ravel()
print(f"TP={tp} FP={fp} FN={fn} TN={tn}")
print(f"ROC-AUC   : {roc_auc_score(y_val, y_proba_naive):.4f}")
print(f"PR-AUC    : {average_precision_score(y_val, y_proba_naive):.4f}")
# TP will likely be 0 — the entire lesson in one number

# ================================================
# STEP 3 — Fixed model with class_weight
# ================================================
fixed = LogisticRegression(
    class_weight='balanced',
    max_iter=1000
)
fixed.fit(X_tr_s, y_train)
y_pred_fixed  = fixed.predict(X_v_s)
y_proba_fixed = fixed.predict_proba(X_v_s)[:, 1]

print("\n=== FIXED MODEL (class_weight='balanced') ===")
print(classification_report(y_val, y_pred_fixed, digits=4))
tn, fp, fn, tp = confusion_matrix(y_val, y_pred_fixed).ravel()
print(f"TP={tp} FP={fp} FN={fn} TN={tn}")
print(f"ROC-AUC   : {roc_auc_score(y_val, y_proba_fixed):.4f}")
print(f"PR-AUC    : {average_precision_score(y_val, y_proba_fixed):.4f}")

# ================================================
# STEP 4 — SMOTE inside pipeline
# ================================================
pipe_smote = ImbPipeline([
    ('scaler', StandardScaler()),
    ('smote',  SMOTE(k_neighbors=5, random_state=42)),
    ('clf',    LogisticRegression(max_iter=1000))
])

cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
scores_smote = cross_val_score(
    pipe_smote, X_train, y_train,
    cv=cv, scoring='average_precision', n_jobs=-1
)
print(f"\n=== SMOTE PIPELINE CV ===")
print(f"PR-AUC per fold : {scores_smote.round(4)}")
print(f"Mean ± Std      : {scores_smote.mean():.4f} ± {scores_smote.std():.4f}")

# ================================================
# STEP 5 — Compare all resampling strategies
# ================================================
strategies = {
    'class_weight only':  ImbPipeline([
        ('scaler', StandardScaler()),
        ('clf',    LogisticRegression(
                       class_weight='balanced', max_iter=1000))
    ]),
    'smote':              ImbPipeline([
        ('scaler',  StandardScaler()),
        ('sampler', SMOTE(random_state=42)),
        ('clf',     LogisticRegression(max_iter=1000))
    ]),
    'adasyn':             ImbPipeline([
        ('scaler',  StandardScaler()),
        ('sampler', ADASYN(random_state=42)),
        ('clf',     LogisticRegression(max_iter=1000))
    ]),
    'borderline_smote':   ImbPipeline([
        ('scaler',  StandardScaler()),
        ('sampler', BorderlineSMOTE(random_state=42)),
        ('clf',     LogisticRegression(max_iter=1000))
    ]),
    'undersample':        ImbPipeline([
        ('scaler',  StandardScaler()),
        ('sampler', RandomUnderSampler(random_state=42)),
        ('clf',     LogisticRegression(max_iter=1000))
    ]),
    'smote_tomek':        ImbPipeline([
        ('scaler',  StandardScaler()),
        ('sampler', SMOTETomek(random_state=42)),
        ('clf',     LogisticRegression(max_iter=1000))
    ]),
}

print("\n=== RESAMPLING STRATEGY COMPARISON ===")
for name, pipe in strategies.items():
    score = cross_val_score(
        pipe, X_train, y_train,
        cv=cv, scoring='average_precision', n_jobs=-1
    ).mean()
    print(f"{name:25s}: PR-AUC = {score:.4f}")

# ================================================
# STEP 6 — Compute class weights manually
# ================================================
weights = compute_class_weight(
    'balanced', classes=np.unique(y_train), y=y_train)
weight_dict = {0: weights[0], 1: weights[1]}
print(f"\nComputed class weights: {weight_dict}")

cost_ratio = 100
manual_weights = {0: 1, 1: cost_ratio}
clf_cost = LogisticRegression(
    class_weight=manual_weights, max_iter=1000)
clf_cost.fit(X_tr_s, y_train)
y_proba_cost = clf_cost.predict_proba(X_v_s)[:, 1]
print(f"Cost-ratio model PR-AUC: "
      f"{average_precision_score(y_val, y_proba_cost):.4f}")

# ================================================
# STEP 7 — Threshold sweep
# ================================================
y_proba = y_proba_fixed
precisions, recalls, thresholds = precision_recall_curve(
    y_val, y_proba)

f1_scores = (2 * precisions * recalls /
             (precisions + recalls + 1e-9))
best_idx       = np.argmax(f1_scores[:-1])
best_threshold = thresholds[best_idx]

print(f"\n=== THRESHOLD SWEEP ===")
print(f"Default 0.5 F1   : {f1_score(y_val, y_proba>=0.5):.4f}")
print(f"Optimal threshold: {best_threshold:.4f}")
print(f"Optimal F1       : {f1_scores[best_idx]:.4f}")

y_pred_optimal = (y_proba >= best_threshold).astype(int)
print(classification_report(y_val, y_pred_optimal, digits=4))

# ================================================
# STEP 8 — PR Curve plot
# ================================================
plt.figure(figsize=(8, 5))
plt.plot(recalls, precisions,
         label=f'PR Curve (AUC={average_precision_score(y_val, y_proba):.4f})')
plt.axvline(recalls[best_idx], color='red', linestyle='--',
            alpha=0.7, label=f'Optimal threshold={best_threshold:.3f}')
plt.xlabel('Recall (Fraud Catch Rate)')
plt.ylabel('Precision (Flag Accuracy)')
plt.title('Precision-Recall Curve')
plt.legend()
plt.tight_layout()
plt.show()

# ================================================
# STEP 9 — The fraud report function
# build this reflex — call it on every single model
# ================================================
def fraud_report(y_true, y_pred, y_proba, label="Model"):
    print(f"\n{'='*55}")
    print(f"  {label}")
    print(f"{'='*55}")
    print(classification_report(y_true, y_pred, digits=4))
    print(f"ROC-AUC  : {roc_auc_score(y_true, y_proba):.4f}")
    print(f"PR-AUC   : {average_precision_score(y_true, y_proba):.4f}")
    tn, fp, fn, tp = confusion_matrix(y_true, y_pred).ravel()
    print(f"TP={tp:5d} | FP={fp:5d} | FN={fn:5d} | TN={tn:7d}")
    print(f"Catch Rate (Recall) : {tp/(tp+fn):.4f}")
    print(f"False Alarm Rate    : {fp/(fp+tn):.6f}")

fraud_report(y_val, y_pred_naive,   y_proba_naive,   "Naive Model")
fraud_report(y_val, y_pred_fixed,   y_proba_fixed,   "class_weight Fixed")
fraud_report(y_val, y_pred_optimal, y_proba,         "Optimal Threshold")
```
